```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
def generate_tartan(fig, ax):
    stripes = np.random.randint(4, 10)
    potential_indices = np.arange(250)
    rand_indices = np.sort(np.random.choice(potential_indices, stripes, replace=False))

    bool = 1
    booleans = []
    for i in range(250):
        if i in rand_indices:
            bool ^= 1
        booleans.append(bool)

    ax.set_facecolor(np.random.rand(3,))
    linewidth = 0.75
    color = np.random.rand(3,)
    for i in range(250):
        if booleans[i]==1:
            ax.axhline(i/1000.0, 0, 1, linewidth=linewidth, color=color, linestyle='dotted')
            ax.axhline(1-i/1000.0, 0, 1, linewidth=linewidth, color=color, linestyle='dotted')
            ax.axvline(i/1000.0, 0, 1, linewidth=linewidth, color=color, linestyle='dotted')
            ax.axvline(1-i/1000.0, 0, 1, linewidth=linewidth, color=color, linestyle='dotted')
        else:
            color = np.random.rand(3,)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

generate_tartan(fig, ax)
```
